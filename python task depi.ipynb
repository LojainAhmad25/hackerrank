```python
if __name__ == '__main__':
    students = [] 
    
for _ in range(int(input())):
        name = input()
        score = float(input())
        students.append([name, score])  
    
   
    scores = sorted(set(student[1] for student in students))
    second_lowest_score = scores[1]  
    
    
    second_place_names = []  
    
    for student in students:
        if student[1] == second_lowest_score:
            second_place_names.append(student[0])
    
    second_place_names.sort()
    
    for name in second_place_names:
        print(name)
```

```python
import math
import os
import random
import re
import sys
from datetime import datetime

# Complete the time_delta function below.
def time_delta(t1, t2):
   
    time_format = "%a %d %b %Y %H:%M:%S %z"
    
    dt1 = datetime.strptime(t1, time_format)
    dt2 = datetime.strptime(t2, time_format)
    
    diff_seconds = abs((dt1 - dt2).total_seconds())
    
    return str(int(diff_seconds))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input().strip())

    for t_itr in range(t):
        t1 = input().strip()
        t2 = input().strip()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re
import sys

input = sys.stdin.read
data = input().splitlines()

S = data[0]
k = data[1]

pattern = re.escape(k)
regex = re.compile(pattern)

matches_found = False
pos = 0

while True:
    m = regex.search(S, pos)
    if not m:
        break
    matches_found = True
    start_idx = m.start()
    end_idx = m.end() - 1
    print("(%d, %d)" % (start_idx, end_idx))
    pos = start_idx + 1

if not matches_found:
    print("(-1, -1)")   
```
